```python
import pandas as pd
corp = pd.read_csv("word_complete.csv", encoding="utf-8-sig")
corp.head()
```

```python
# 빈도 확인
fre = corp["keyword"]
corpus = [str(i).split(" ") for i in fre]
```

```python
def word_count(corpus): # 워드카운트 함수 
    word_cnt = dict()
    for text in corpus:
        for word in text:
            if word_cnt.get(word):
                word_cnt[word]+=1 #단어를 가져오면 카운트 아니면 그대로 진행
            else:
                word_cnt[word]=1
    cnt_df = pd.DataFrame.from_dict(word_cnt, orient='index')
    cnt_df.columns = ["count"]
    cnt_df = cnt_df.sort_values(["count"],ascending=False)
    return word_cnt, cnt_df
```

```python
words_cnt, cnt_df = word_count(corpus)
```

```python
cnt_df.head()
```

```python
cnt_df = cnt_df[cnt_df['count']>100]
```

```python
cnt_df.to_csv("word_fre.csv", encoding="utf-8-sig", index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
from sklearn.decomposition import LatentDirichletAllocation
import pickle
from gensim import corpora
import gensim
```

```python
# 토픽별 단어 분포

dictionary = corpora.Dictionary(corpus)
corpus = [dictionary.doc2bow(text) for text in corpus] #단어 아이디, 해당단어 빈도수 튜플
NUM_TOPICS = 5 # 토픽 5개 설정 
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15) # 15회 실시 에포크와 동일
pickle.dump(corpus, open('corpus_pre.pkl', 'wb'))
```

```python
ldamodel.show_topic(0) # 주제분류 결과는 해당 토픽과 가중치가 높은 10개 단어
```

```python
import pyLDAvis
import pyLDAvis.gensim_models
pyLDAvis.enable_notebook() # 
vis = pyLDAvis.gensim_models.prepare(ldamodel, corpus, dictionary) # 시각화를 위해 LDA 모델의 데이터 변환 및 준비
pyLDAvis.save_html(vis,"lda_pre.html") #html로 저장
pyLDAvis.display(vis)
```
