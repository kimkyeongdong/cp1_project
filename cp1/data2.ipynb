#### 통계청 데이터

* 큰 가공은 필요없으나 DB로 변환

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("1일_평균_방문.csv", encoding='cp949')
df.head(30)
df.columns = ['전체','전체2','전체3','전체4','시점','방문배달포장', '방문배달포장인원','원데이터','전년대비증감','증감률']
# 시점은 2021 시점으로 통계청 자료에 아직 2022 자료가 없음
# '전체': 업종별, 지역별, 운영형태별
# '전체2': 일반음식점, 일반음식점 외, 프랜차이즈, 비프랜차이즈, 지역별 
# '전체3': 한식,일식, 기관내내식당, 주점 등등
# '전체4': 기타음식점 분류
```

```python
df.info()
```

```python
df = df.fillna(0)
df.info()
```

```python
df.head(20)
```

# DB변환 및 확인

```python
import sqlite3
connection = sqlite3.connect('eat_out.db')
```

```python
df.to_sql('day_mean',connection,if_exists='replace',index=False) # append값 추가가능 기본적으로 fail
connection.commit()
connection.close()   
```

```python
connection = sqlite3.connect('eat_out.db')
df3 = pd.read_sql("SELECT * FROM day_mean;",connection)
connection.commit()
connection.close() 
```
